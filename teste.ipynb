```python
import os
from dotenv import load_dotenv
import requests

load_dotenv()  # carrega o .env da raiz
GFW_API_TOKEN = os.environ.get("GFW_API_TOKEN")
dataset = "wri_tropical_tree_cover_extent"
data = "0220922"

if GFW_API_TOKEN:
    url = ("https://data-api.globalforestwatch.org/dataset/{dataset}/v2{data}/download/geotiff"        
        "?grid=10/100000&tile_id=00N_000E&pixel_meaning=decile")
    headers = {"Authorization": f"Bearer {GFW_API_TOKEN}"}

    try:
        r = requests.get(url, headers=headers, timeout=180)
        r.raise_for_status()
        with open("ttc_tile_00N_000E.tif", "wb") as f:
            f.write(r.content)
    except Exception as ex:
        print(ex)
        
    print(" baixado GeoTIFF")
else:
    print("Token no encontrado nas variaveis de ambiente, favor validar!")
```
